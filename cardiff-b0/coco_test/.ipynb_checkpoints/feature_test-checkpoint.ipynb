```python
from quantize_audio_feature import *
import librosa
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
feature_cfg = {
    "n_fft"      : 512,
    "sr"         : 16000,
    "win_length" : 512,
    "hop_length" : 256,
    "n_mels"     : 40,
    "fmin"   : 20,
    "fmax"   : 8000
}
qlist = {
    "win"  : [1,16,15],
    "fft"  : [1,16,10],
    "pow"  : [0,20,10],
    "mel"  : [1,16,15],
    "spec" : [0,16,13] 
}
audio_path = "0c40e715_nohash_0.wav"
y, sr = librosa.load(audio_path, sr=16000)
plt.plot(y)
```

```python
qf = qFeature( qlist = qlist, feature_cfg = feature_cfg)
window = qf.qfft_window
mel_filter = qf.mel_filter_q
# plt.figure()
# plt.plot(window)
# plt.figure()
# plt.imshow(mel_filter)
```

```python
qfbank = qf.fbank_quantize(y,db=True)
print(qfbank.shape)
plt.imshow(qfbank)
```

```python
now = time.time()
librosa_fbank = librosa.feature.melspectrogram(
    y=y,
    sr=feature_cfg["sr"],
    S=None,
    n_fft=feature_cfg["n_fft"],
    hop_length=feature_cfg["hop_length"],
    win_length=feature_cfg["win_length"],
    window="hann",
    center=False,
    pad_mode="reflect",
    power=2.0,
    n_mels = feature_cfg["n_mels"],
    fmin = feature_cfg["mel_fmin"],
    fmax = feature_cfg["mel_fmax"]
)
print('time: {}'.format(time.time()-now))
librosa_fbank_db = librosa.power_to_db(librosa_fbank)
print(librosa_fbank.shape)
plt.imshow(librosa_fbank_db)
```
